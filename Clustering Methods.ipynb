```python
# Importing the metrics libraries to perform internal validation
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from validclust import dunn

# CH ratio
# CH = (SSBm /(M−1))/(SSEm /(M))
# Where M is number of clusters

print(metrics.calinski_harabasz_score(var_task2_df.iloc[:,:19], var_task2_km_labels))
print(metrics.calinski_harabasz_score(var_task2_df.iloc[:,:19], var_task2_hc_labels))

#Dunn Index
#DIm = min1⩽i<j⩽m δ(Ci,Cj) / max1⩽k⩽m ∆k

distances_val = pairwise_distances(var_task2_df.iloc[:,0:19])

dunn_km = dunn(distances_val,  var_km.labels_)
dunn_hc = dunn(distances_val,  var_task2_hierarchical_cluster.labels_)

print(str(dunn_km))
print(str(dunn_hc))
```

```python
# Silhouette score(silhouette mean) for Kmeans
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
silhouette_avg_km = []
for num_clusters in range_n_clusters:
  kmeans = KMeans(n_clusters=num_clusters)
  kmeans.fit(var_task2_df.iloc[:,:19])
  cluster_labels_km = kmeans.labels_
 
 # silhouette score
  silhouette_avg_km.append(silhouette_score(var_task2_df.iloc[:,:19], cluster_labels_km))

print(f"cluster range: {range_n_clusters}")

print(f"silhousette score: {silhouette_avg_km}")

var_plt.plot(range_n_clusters, silhouette_avg_km)
var_plt.xlabel('Values of K') 
var_plt.ylabel('Silhouette score') 
var_plt.title('Silhouette analysis For Optimal k in Kmeans')
var_plt.show()
```

```python
import math
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Visualizing Silhouette score for each cluster in Kmeans Clusting

def visualize_silhouette_kmeans(cluster_lists, X_features):     
    # get cluster_list
    n_cols = len(cluster_lists)
    
    # subplot
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
    
    #
    for ind, n_cluster in enumerate(cluster_lists):
        
        # KMeans clustering, sil_avg for silhouette average and sil_values for each silhouette values. 
        clusterer = KMeans(n_clusters = n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(X_features)
        #print(n_cluster)
        #print(cluster_labels)
        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)
        
        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
        
        # For graph 
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels==i]
            ith_cluster_sil_values.sort()
            
            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i
            
            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10
            
        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")

visualize_silhouette_kmeans([2, 3, 4, 5, 6, 7, 8, 9, 10], var_task2_df.iloc[:,:19])
```

```python
import math
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np

from sklearn.metrics import silhouette_samples, silhouette_score

# Visualizing Silhouette score for each cluster in Hierarchial Clusting

def visualize_silhouette_hierarchical(cluster_lists, X_features):     
    # get cluster_list
    n_cols = len(cluster_lists)
    
    # subplot
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
    
    #
    for ind, n_cluster in enumerate(cluster_lists):
        
        hc_cluster = AgglomerativeClustering(n_clusters= n_cluster, affinity='euclidean', linkage='ward')
        hc_labels = hc_cluster.fit_predict(X_features)
        #print(n_cluster)
        #print(hc_labels)

        sil_avg = silhouette_score(X_features, hc_labels)
        sil_values = silhouette_samples(X_features, hc_labels)
        
        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
        
        # graph
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[hc_labels==i]
            ith_cluster_sil_values.sort()
            
            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i
            
            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10
            
        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")

visualize_silhouette_hierarchical([2, 3, 4, 5, 6, 7, 8, 9, 10], var_task2_df.iloc[:,:19])
```
